```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])
    
df["year"] = years
```

With your dataframe set up, time to start creating visualizations!

```python
df.columns
```

```python
num_books_per_pub = df.groupby('publisher')['title'].count().sort_values(ascending=False)
num_books_per_pub
```

```python
# Use your Python skills to create a visualization here.
df_top10pub = pd.DataFrame(num_books_per_pub.nlargest(25))

df_top10pub.plot.bar(legend = False)
plt.title('Top 50 Publishers (by number of books published)',color = 'white')
plt.xlabel('Publisher')
plt.ylabel('Number of books')
plt.title('Number of books published by each publisher')
plt.show()
```

```python
# Use your Python skills to create a visualization here.
num_books_per_pub_lang= df.groupby('publisher')['title'].count().sort_values(ascending=False)
num_books_per_pub_lang
```

```python
counts = df.groupby('publisher')['title'].count().sort_values( ascending=False)[:25]
df_top25pub = pd.DataFrame(counts)
print(counts)
sb.barplot(data=df_top25pub, x=df_top25pub.index, y='title')
plt.xticks(rotation=90)
plt.xlabel('Publisher')
plt.ylabel('Number of books')
plt.title('Number of books published by each publisher')
plt.show()
```
